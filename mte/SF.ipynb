```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, gStyle, TFile, TIter, gSystem, TTree, TH1F, TPaveStats, TGraph, TH2F, TProfile
import sys
gROOT.SetStyle()
gROOT.SetBatch(True)
from array import array
import mcCorrections
import numpy
```

```python
f = TFile('../../FinalStateAnalysis/TagAndProbe/data/htt_scalefactors_v17_5.root')
w = f.Get("w")
tEff = mcCorrections.efficiency_trigger_mu_2017

canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.52,0.6,0.9,0.7)
h = TH1F("h","",500,0,500)
h1 = TH1F("h1","",500,0,500)
h1.GetXaxis().SetTitle("m Pt (GeV)")
h1.SetLineColor(2)
                                   
for i in range(500):
    w.var("m_pt").setVal(i)
    w.var("m_eta").setVal(0)
    h.Fill(i, w.function("m_trg27_embed_kit_ratio").getVal())
    h1.Fill(i, tEff(i, 0))  
                                                                          
h1.Draw("hist")
h.Draw("hist SAME")
legend.AddEntry("h", "HTT Embed Muon Trigger SF")
legend.AddEntry("h1", "Central Muon Trigger SF")
legend.Draw()
canvas.Draw()
```

```python
f = TFile('../../FinalStateAnalysis/TagAndProbe/data/htt_scalefactors_2017_v1.root')
w = f.Get("w")

canvas = TCanvas("canvas","canvas",800,800)
h = TH1F("h","",50,-2.5,2.5)
h.GetXaxis().SetTitle("m Eta")
                                   
for i in numpy.arange(-2.4, 2.4, 0.1):
    w.var("e_pt").setVal(50)
    w.var("e_eta").setVal(i)
    w.var("m_pt").setVal(50)                                                                                                                                                                                                                                          
    w.var("m_eta").setVal(i)
    tData = w.function("m_trg_binned_23_data").getVal()*w.function("e_trg_binned_12_data").getVal()                                                                                                                                                                       
    tEmbed = w.function("m_trg_binned_23_embed").getVal()*w.function("e_trg_binned_12_embed").getVal()
    h.Fill(i, tData/tEmbed)
                                                                          
h.Draw("hist")
canvas.Draw()
```
