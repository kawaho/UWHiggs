```python
import os
from sys import argv, stdout, stderr
import ROOT
import sys                                                                                                                                                                                                                                                  
ROOT.gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
ROOT.gROOT.SetBatch(True)
```

```python
fdata = ROOT.TFile('Data.root')
canvas = ROOT.TCanvas("canvas","canvas",1200,1200)
canvas.Divide(3,3)
canvas.cd(1)
h = fdata.Get('BDT/ePt')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(2)
h = fdata.Get('BDT/tPt')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(3)
h = fdata.Get('BDT/dPhiETau')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(4)
h = fdata.Get('BDT/dEtaETau')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(5)
h = fdata.Get('BDT/e_t_collinearMass')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(6)
h = fdata.Get('BDT/e_t_visibleMass')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(7)
h = fdata.Get('BDT/MTTauMET')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.cd(8)
h = fdata.Get('BDT/dPhiTauMET')
h.SetMinimum(0)
h.Draw("COLZ")
canvas.Draw()
```
