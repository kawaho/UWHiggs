```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
def positivize(histogram):                                                                                                                                                                                                                                                      
    output = histogram.Clone()                                                                                                                                                                                                                                                  
    for i in range(output.GetSize()):                                                                                                                                                                                                                                           
        if output.GetArray()[i] < 0:                                                                                                                                                                                                                                            
            output.AddAt(0, i)                                                                                                                                                                                                                                                  
    return output
```

```python
canvas = TCanvas("canvas","canvas",800,800)
fdata = TFile('QCD_Pt-20toInf_MuEnrichedPt15_TuneCP5_13TeV_pythia8_v14-v1.root')
femb = TFile('embedded_EmbeddingRun2017B_ElTauFinalState.root')
fmc = TFile('MC.root')

hdata = fdata.Get('TightOS0Jet/ePt')
hemb = femb.Get('TightOS0Jet/ePt')
hembt = femb.Get('TauLooseOS0Jet/ePt')
hembe = femb.Get('EleLooseOS0Jet/ePt')
hembet = femb.Get('EleLooseTauLooseOS0Jet/ePt')
hembf = hembt.Clone()
hembf.Add(hembe)
hembf.Add(hembet, -1)
hembf = positivize(hembf)
he = hemb.Clone()
he.Add(hembf, -1)
htau = fdata.Get('TauLooseOS0Jet/ePt')
hele = fdata.Get('EleLooseOS0Jet/ePt')
het = fdata.Get('EleLooseTauLooseOS0Jet/ePt')
hf = htau.Clone()
hf.Add(hele)
hf.Add(het, -1)
hmc = fmc.Get('TightOS0Jet/ePt')
hmct = fmc.Get('TauLooseOS0Jet/ePt')
hmce = fmc.Get('EleLooseOS0Jet/ePt')
hmcet = fmc.Get('EleLooseTauLooseOS0Jet/ePt')
hmcf = hmct.Clone()
hmcf.Add(hmce)
hmcf.Add(hmcet, -1)
hmcf = positivize(hmcf)
hm = hmc.Clone()
hm.Add(hmcf, -1)
hb = hm.Clone()
hb.Add(hf)
hed0 = hdata.Clone()
hed0.Add(hb, -1)
hed0.Divide(he)

hdata = fdata.Get('TightOS1Jet/ePt')
hemb = femb.Get('TightOS1Jet/ePt')
hembt = femb.Get('TauLooseOS1Jet/ePt')
hembe = femb.Get('EleLooseOS1Jet/ePt')
hembet = femb.Get('EleLooseTauLooseOS1Jet/ePt')
hembf = hembt.Clone()
hembf.Add(hembe)
hembf.Add(hembet, -1)
hembf = positivize(hembf)
he = hemb.Clone()
he.Add(hembf, -1)
htau = fdata.Get('TauLooseOS1Jet/ePt')
hele = fdata.Get('EleLooseOS1Jet/ePt')
het = fdata.Get('EleLooseTauLooseOS1Jet/ePt')
hf = htau.Clone()
hf.Add(hele)
hf.Add(het, -1)
hmc = fmc.Get('TightOS1Jet/ePt')
hmct = fmc.Get('TauLooseOS1Jet/ePt')
hmce = fmc.Get('EleLooseOS1Jet/ePt')
hmcet = fmc.Get('EleLooseTauLooseOS1Jet/ePt')
hmcf = hmct.Clone()
hmcf.Add(hmce)
hmcf.Add(hmcet, -1)
hmcf = positivize(hmcf)
hm = hmc.Clone()
hm.Add(hmcf, -1)
hb = hm.Clone()
hb.Add(hf)
hed1 = hdata.Clone()
hed1.Add(hb, -1)
hed1.Divide(he)

hdata = fdata.Get('TightOS2Jet/ePt')
hemb = femb.Get('TightOS2Jet/ePt')
hembt = femb.Get('TauLooseOS2Jet/ePt')
hembe = femb.Get('EleLooseOS2Jet/ePt')
hembet = femb.Get('EleLooseTauLooseOS2Jet/ePt')
hembf = hembt.Clone()
hembf.Add(hembe)
hembf.Add(hembet, -1)
hembf = positivize(hembf)
he = hemb.Clone()
he.Add(hembf, -1)
htau = fdata.Get('TauLooseOS2Jet/ePt')
hele = fdata.Get('EleLooseOS2Jet/ePt')
het = fdata.Get('EleLooseTauLooseOS2Jet/ePt')
hf = htau.Clone()
hf.Add(hele)
hf.Add(het, -1)
hmc = fmc.Get('TightOS2Jet/ePt')
hmct = fmc.Get('TauLooseOS2Jet/ePt')
hmce = fmc.Get('EleLooseOS2Jet/ePt')
hmcet = fmc.Get('EleLooseTauLooseOS2Jet/ePt')
hmcf = hmct.Clone()
hmcf.Add(hmce)
hmcf.Add(hmcet, -1)
hmcf = positivize(hmcf)
hm = hmc.Clone()
hm.Add(hmcf, -1)
hb = hm.Clone()
hb.Add(hf)
hed2 = hdata.Clone()
hed2.Add(hb, -1)
hed2.Divide(he)

f = TFile('Pt.root',"RECREATE")
f.cd()
hed0.SetName('0Jet')
hed1.SetName('1Jet')
hed2.SetName('2Jet')
hed0.Write()
hed1.Write()
hed2.Write()
f.Close()
```
