```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
canvas = TCanvas("canvas","canvas",800,800)

fdy = TFile('DY.root')
fdata = TFile('data.root')
fdb = TFile('Diboson.root')

hdy = fdy.Get('initial/m1_m2_Mass')
hdata = fdata.Get('initial/m1_m2_Mass')
hdb = fdb.Get('initial/m1_m2_Mass')

hqcd = hdata.Clone()
hqcd.Add(hdb, -1)
hqcd.Divide(hdy)

hqcd.Draw()
canvas.Draw()

print hqcd.GetBinContent(3)
print hqcd.GetBinContent(4)
print hqcd.GetBinContent(5)
print hqcd.GetBinContent(6)
```
