```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, gStyle, TFile, TIter, gSystem, TTree, TH1F, TPaveStats, TGraph, TH2F, TProfile
import sys
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
from array import array
```

```python
f = TFile('QCD_Pt-20toInf_MuEnrichedPt15_TuneCP5_13TeV_pythia8_v14-v1.root')
ht = f.Get('TauLooseOS/m_t_CollinearMass')
hm = f.Get('MuonLooseOS/m_t_CollinearMass')
hmt = f.Get('MuonLooseTauLooseOS/m_t_CollinearMass')
h = ht.Clone()
h.Add(hm)
h.Add(hmt)
h.SetLineColor(2)
canvas = TCanvas("canvas","canvas",800,800)
h.Draw()
ht.Draw("SAME")
canvas.Draw()
```

```python
ht = f.Get('TauLooseOS0Jet/m_t_CollinearMass')
hm = f.Get('MuonLooseOS0Jet/m_t_CollinearMass')
hmt = f.Get('MuonLooseTauLooseOS0Jet/m_t_CollinearMass')
h = ht.Clone()
h.Add(hm)
h.Add(hmt)
h.SetLineColor(2)
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.52,0.6,0.8,0.7)
h.Draw()
ht.Draw("SAME")
legend.AddEntry(ht, "Tau Fakes")
legend.AddEntry(h, "Muon+Tau Fakes")
legend.Draw()
canvas.Draw()
```
