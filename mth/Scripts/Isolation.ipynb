```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.52,0.7,0.8,0.8)
fe = TFile('../results/DataApril3/AnalyzeMuTauGen/embed.root')                                                                                                                                                                                                                              
hte = fe.Get('TightOS/mPt')
h1 = hte.Clone()
hme = fe.Get('MuonLooseOS/mPt')
h1.Divide(hme)

fd = TFile('../results/DataApril3/AnalyzeMuTauGen/Drell.root')                                                                                                                                                                                                                              
htd = fd.Get('TightOS/mPt')
h2 = htd.Clone()
hmd = fd.Get('MuonLooseOS/mPt')
h2.Divide(hmd)
h2.SetLineColor(2)

legend.AddEntry(h1, "Embed")
legend.AddEntry(h2, "Drell")

h1.SetTitle("Muon Isolation T&P")
h1.Draw()
h2.Draw('SAME')
legend.Draw()
canvas.Draw()
```
