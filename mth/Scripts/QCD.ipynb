```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.52,0.7,0.8,0.8)
f1 = TFile('../results/DataApril3/AnalyzeMuTauZTT/QCD_Pt-20toInf_MuEnrichedPt15_TuneCP5_13TeV_pythia8_v14-v1.root')                                                                                                                                                                                                                              
h1 = f1.Get('TauLooseOS/m_t_CollinearMass')
hf = h1.Clone()
hf.SetLineColor(2)
h2 = f1.Get('MuonLooseOS/m_t_CollinearMass')
h3 = f1.Get('MuonLooseTauLooseOS/m_t_CollinearMass')
hf.Add(h2)
hf.Add(h3, -1)
hf.Draw()
h1.Draw('SAME')
legend.AddEntry(h1, "Tau Loose")
legend.AddEntry(hf, "Total Fakes")
print h1.Integral()
print h2.Integral()
print h3.Integral()
print hf.Integral()
legend.Draw()
canvas.Draw()
```
