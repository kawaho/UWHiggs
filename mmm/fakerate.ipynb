```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, TFile, gPad, TPad, TF1, gStyle
import sys
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
```

```python
gStyle.SetOptFit(1)
canvas = TCanvas('canvas','canvas',800,800)
grid = TPad("grid", "", 0, 0, 1, 1)
grid.Draw()
grid.cd()
fi = TFile('frMuonPt.root')
h = fi.Get('fakerate')
h.Draw()
gPad.Update()
graph = h.GetPaintedGraph()
h.SetTitle("Muon Fake Rate; Muon pT (GeV); Fake Rate")
graph.SetMinimum(0.5)
graph.SetMaximum(1.5)
gPad.Update()
f = TF1("Linear law","[0]+x*[1]", 0, 200)
f.SetLineColor(4)
f.SetLineStyle(0)
h.Fit(f)
f.DrawClone("Same")
canvas.Draw()
```

```python
gStyle.SetOptFit(1)
canvas = TCanvas('canvas','canvas',800,800)
grid = TPad("grid", "", 0, 0, 1, 1)
grid.Draw()
grid.cd()
fi = TFile('frMuonEta.root')
h = fi.Get('fakerate')
h.Draw()
gPad.Update()
graph = h.GetPaintedGraph()
h.SetTitle("Muon Fake Rate; Muon Eta; Fake Rate")
graph.SetMinimum(0.5)
graph.SetMaximum(1.0)
gPad.Update()
f = TF1("Linear law","[0]+x*[1]", -2.4, 2.4)
f.SetLineColor(4)
f.SetLineStyle(0)
h.Fit(f)
f.DrawClone("Same")
canvas.Draw()
```
