```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, gStyle, TFile, TIter, gSystem, TTree
import sys
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
gStyle.SetOptStat(0)
```

```python
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.6,0.7,0.89,0.89)
file = 'TMVA.root'
MuTaufile = TFile(file)
hS = MuTaufile.Get('dataset/Method_BDT/BDT/MTTauMET__Signal')
hS.GetXaxis().SetRangeUser(0, 200)
hS.GetYaxis().SetLabelSize(0.025)
hS.SetFillStyle(3001)
hS.SetFillColor(2)
hB = MuTaufile.Get('dataset/Method_BDT/BDT/MTTauMET__Background')
hB.GetXaxis().SetRangeUser(0, 200)
hB.GetYaxis().SetLabelSize(0.025)
hB.SetFillStyle(3002)
hB.SetFillColor(3)
hS.Draw('hist')
hB.Draw('hist SAME')
legend.AddEntry(hS, "Signal")
legend.AddEntry(hB, "Background")
legend.Draw()
canvas.Draw()
canvas.SaveAs('MTTauMET.png')
```

```python
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.3,0.7,0.59,0.89)
file = '../TMVA.root'
MuTaufile = TFile(file)
hS = MuTaufile.Get('dataset/Method_BDT/BDT/MVA_BDT_S')
hST = MuTaufile.Get('dataset/Method_BDT/BDT/MVA_BDT_Train_S')
hS.SetTitle('TMVA overtraining check for classifier: BDT')
hS.SetLineColor(2)
hST.SetFillStyle(3001)
hST.SetFillColor(2)
hB = MuTaufile.Get('dataset/Method_BDT/BDT/MVA_BDT_B')
hBT = MuTaufile.Get('dataset/Method_BDT/BDT/MVA_BDT_Train_B')
hB.SetLineColor(3)
hBT.SetFillStyle(3002)
hBT.SetFillColor(3)
hS.Draw('e')
hB.Draw('e SAME')
hST.Draw('hist SAME')
hBT.Draw('hist SAME')
legend.AddEntry(hS, "Signal Train")
legend.AddEntry(hB, "Background Train")
legend.AddEntry(hST, "Signal Test")
legend.AddEntry(hBT, "Background Test")
legend.Draw()
canvas.Draw()
canvas.SaveAs('TMVAch.png')
```
