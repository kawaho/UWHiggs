```python
canvas = TCanvas("canvas","canvas",800,800)
file = 'pileupQCD.root'
f = TFile("pileupQCDnew.root","RECREATE")
MuTaufile = TFile(file)
lfv_histo = MuTaufile.Get('pileup')
hlist = lfv_histo.GetListOfKeys()
iter = TIter(hlist)
for i in iter:
    if i.GetName()=='pileup':
        histo = MuTaufile.Get('pileup/'+i.GetName())
        print histo.Integral()
        histo.Draw("hist")
canvas.Draw()
histo.SetDirectory(f)
f.Write()
```
