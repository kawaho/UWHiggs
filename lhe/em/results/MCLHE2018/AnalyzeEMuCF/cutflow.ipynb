```python
from ROOT import TFile, TIter, TCanvas, gROOT, gStyle
gROOT.SetStyle("Default")
gROOT.SetBatch(True)
gStyle.SetOptTitle(0)
```

```python
canvas = TCanvas("canvas","canvas",800,800)

f = "VBF_LFV_HToETau_M125_TuneCP5_PSweights_13TeV_powheg_pythia8_-102X_upgrade2018_realistic_v15-v1.root"

MuTaufile = TFile(f)
cutflowhisto = MuTaufile.Get('./CUT_FLOW')
cutflowhisto.Draw("hist")
canvas.Draw()
```

```python
print "Overall Cut Efficiency: ", round(cutflowhisto.GetBinContent(13)/cutflowhisto.GetBinContent(1), 3)
```
