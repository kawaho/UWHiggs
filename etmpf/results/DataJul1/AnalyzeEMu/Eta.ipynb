```python
import os
from sys import argv, stdout, stderr
from ROOT import TFile, gROOT, TCanvas, TLegend
import sys                                                                                                                                                                                                                                                  
gROOT.SetStyle("Plain")                                                                                                                                                                                                                                           
gROOT.SetBatch(True)
```

```python
canvas = TCanvas("canvas","canvas",800,800)
f1 = TFile('GluGlu.root')
f2 = TFile('GluGlu_LFV_HToETau_M125_13TeV_powheg_pythia8_v2-PU2017_12Apr2018_94X_mc2017_realistic_v14-v1.root')
h1 = f1.Get('TightOS/e_m_CollMass')
h2 = f2.Get('TightOS/e_m_CollMass')
h2.SetLineColor(3)
h1.Draw()
h2.Draw("SAME")
canvas.Draw()
print h1.Integral()
print h2.Integral()
```
